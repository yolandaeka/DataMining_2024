```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load dataset
df = pd.read_csv('diabetes.csv')

# Normalize data
target_column = ['Outcome']
predictors = list(set(list(df.columns)) - set(target_column))
df[predictors] = df[predictors] / df[predictors].max()

# Prepare data for training and testing
X = df[predictors].values
y = df[target_column].values.ravel()  # flatten y array

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)

# List of activation functions to test
activations = ['identity', 'logistic', 'tanh', 'relu']
results = {}

for activation in activations:
    # Initialize and train the model
    mlp = MLPClassifier(hidden_layer_sizes=(7, 10, 8), activation=activation, solver='adam', max_iter=500)
    mlp.fit(x_train, y_train)

    # Predictions
    predict_train = mlp.predict(x_train)
    predict_test = mlp.predict(x_test)

    # Evaluate the model
    train_accuracy = accuracy_score(y_train, predict_train)
    test_accuracy = accuracy_score(y_test, predict_test)

    results[activation] = {
        'train_accuracy': train_accuracy,
        'test_accuracy': test_accuracy,
        'confusion_matrix': confusion_matrix(y_test, predict_test),
        'classification_report': classification_report(y_test, predict_test)
    }

# Print results
for activation, result in results.items():
    print(f"Activation Function: {activation}")
    print(f"Train Accuracy: {result['train_accuracy']}")
    print(f"Test Accuracy: {result['test_accuracy']}")
    print("Confusion Matrix:")
    print(result['confusion_matrix'])
    print("Classification Report:")
    print(result['classification_report'])
    print("--------------------------------------------------------")
```
